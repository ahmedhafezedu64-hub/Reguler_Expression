```python
import re 
```

```python
# read Data
with open("Data.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
def find_pattern (pattern , Data) : 
    pattern = re.compile(pattern ,re.MULTILINE)
    matches = pattern.finditer(Data)
    return matches
```

```python
def store_data(matches) : 
    founded_list = []
    for match in matches :
        founded_list.append(match.group(0))
    return founded_list
```

```python
phone_pattern =  r"[+20]?\d{11}" 
phone_number_matches = find_pattern ( phone_pattern  , content)
phone_list = store_data (phone_number_matches)
```

```python
email_pattern = r"[a-zA-Z0-9-._]+@[a-zA-Z]+\.[A-Z-a-z]+"
email_Matches = find_pattern(email_pattern , content)
email_list = store_data(email_Matches)
```

```python
RemovePattern = r"({0})|({1})|([ \t*-]+)".format(email_pattern ,phone_pattern)
NameList = [] 
for line in content.split("\n"):
    if line.strip() :
        #   (RemovePattern , Replace , content ,  flags = ignore case  ) Remove the white space between text 
        name = re.sub(RemovePattern , " " , line , flags = re.IGNORECASE).strip()
        name_segment = re.sub(r'\s+', ' ', name)
        if name_segment and len(name_segment.split()) >= 2:
            NameList.append(name_segment.title())

            
NameList
        
    
```

```python
import pandas as pd 
```

```python
df = pd.DataFrame({
    'Full Name': NameList,
     'Email': email_list,
    'phone Number': phone_list
})
df.to_csv('Contact_Data.csv', index=False, encoding='utf-8')
```
